```python
import cv2

# 1) Kamerayı aç (0 = varsayılan kamera)
cap = cv2.VideoCapture(1)

if not cap.isOpened():
    print("Kamera bulunamadı!")
    exit()

while True:
    # 2) Kareleri oku
    ret, frame = cap.read()
    if not ret:
        print("Kare okunamadı!")
        break

    # 3) Görüntüyü göster
    cv2.imshow("Kamera - 'q' ile cikis", frame)

    # 4) Çıkış için 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 5) Kaynakları serbest bırak
cap.release()
cv2.destroyAllWindows()
```
