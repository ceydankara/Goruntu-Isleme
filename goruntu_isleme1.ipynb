```python
import cv2
import matplotlib.pyplot as plt

# ---- 1) Resmi oku ----
# Ltfen kendi bilgisayarndaki bir resmin yolunu yaz (rnek: "C:/Users/LENOVO/Desktop/resim.jpg")
image_path = "resim.jpg"   # <-- buray deitir
img = cv2.imread(image_path)

# OpenCV BGR formatnda okur, matplotlib RGB ister  dntrme
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# ---- 2) Grnty gster ----
plt.figure(figsize=(6,6))
plt.imshow(img_rgb)
plt.title("Orijinal Renkli Grnt")
plt.axis("off")
plt.show()

# ---- 3) Griye evir ----
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(6,6))
plt.imshow(gray, cmap="gray")
plt.title("Gri Grnt")
plt.axis("off")
plt.show()

```

```python
# ---- 4) Hafif bulanklatr (grlty azaltr) ----
blur = cv2.GaussianBlur(gray, (5,5), 0)

# ---- 5) Canny kenar tespiti ----
edges = cv2.Canny(blur, 100, 200)

plt.figure(figsize=(6,6))
plt.imshow(edges, cmap="gray")
plt.title("Canny Kenar Tespiti")
plt.axis("off")
plt.show()
```

```python
cv2.imwrite("gri_goruntu.jpg", gray)
cv2.imwrite("kenarlar.jpg", edges)

print("Resimler baaryla kaydedildi!")
```

```python
import cv2
import matplotlib.pyplot as plt

# ---- 1) Resmi oku ----
image_path = "a.webp"
img = cv2.imread(image_path)

# Dosya bulunmazsa hata ver
if img is None:
    raise FileNotFoundError(f"{image_path} bulunamad. Yol doru mu kontrol et.")

# OpenCV BGR formatnda okur, matplotlib RGB ister  dntrme
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# ---- 2) Orijinal resmi gster ----
plt.figure(figsize=(6,6))
plt.imshow(img_rgb)
plt.title("Orijinal Renkli Grnt")
plt.axis("off")
plt.show()

# ---- 3) Griye evir ----
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(6,6))
plt.imshow(gray, cmap="gray")
plt.title("Gri Grnt")
plt.axis("off")
plt.show()

# ---- 4) Hafif bulanklatr ----
blur = cv2.GaussianBlur(gray, (5,5), 0)

# ---- 5) Canny kenar tespiti ----
edges = cv2.Canny(blur, 100, 200)

plt.figure(figsize=(6,6))
plt.imshow(edges, cmap="gray")
plt.title("Canny Kenar Tespiti")
plt.axis("off")
plt.show()

# ---- 6) ktlar kaydet ----
cv2.imwrite("gri_goruntu.jpg", gray)
cv2.imwrite("kenarlar.jpg", edges)
print("Resimler baaryla kaydedildi!")
```
